```python
import requests
```

```python
url = 'http://openapi.seoul.go.kr:8088/6562434a6d73656f3130304270787075/json/TbParkingInfoView/1/100/'

response = requests.get(url)
```

```python
print(response.content)
```

```python
data = response.json()
data
```

```python
import pandas as pd
temp=[]
for d in data['TbParkingInfoView']['row']:
  temp.append(d)
uni = pd.DataFrame(temp)
uni
```

```python
import folium
anMap = folium.Map(location=[37.88988548,127.1371472],
                    zoom_control=False,
                    zoom_start=11,
                    control_scale= True)
anMap
```

```python


from folium.plugins import MarkerCluster
mc = MarkerCluster().add_to(anMap)

for n, row in uni.iterrows():
  folium.Marker(location=[row['PSTN_INFO_LAT'], row['PSTN_INFO_LOT']], popup = folium.Popup(row['PKLT_TYPE'], max_width=200)).add_to(mc)

anMap
```
