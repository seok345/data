```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster

df = pd.read_excel('/content/자연휴양림출력화일.xlsx')
df.columns = df.columns.str.strip()
print("컬럼명 확인:", df.columns)
possible_name_cols = [col for col in df.columns if '이름' in col or '휴양림' in col]
name_col = possible_name_cols[0] if possible_name_cols else '알 수 없는 휴양림'
latitude_col = '위도'
longitude_col = '경도'
m = folium.Map(location=[df[latitude_col].mean(), df[longitude_col].mean()], zoom_start=11)
marker_cluster = MarkerCluster().add_to(m)
for idx, row in df.iterrows():
    휴양림_이름 = row[name_col] if pd.notna(row[name_col]) else '이름 없음'
    folium.Marker(
        location=[row[latitude_col], row[longitude_col]],
        popup=휴양림_이름,
        tooltip=휴양림_이름
    ).add_to(marker_cluster)
m.save('/content/map.html')
display(m)
```
